```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
%matplotlib inline
```

#### **line plot shows the price trends over time**

```python
plt.figure(figsize = (10,5))
plt.plot(cryptos['CloseTime'] , cryptos['ClosePrice'] , color = 'Brown')
plt.title('Price Trend Analysis')
plt.xlabel('Time')
plt.ylabel('Price')
plt.xticks(rotation = 45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.fill_between(cryptos['CloseTime'] , cryptos['Volume'], color = 'Purple' ) #alpha makes the color transperant
plt.xlabel('Time')
plt.ylabel('Volume')
plt.title('Area Chart of Volume over Time')
plt.grid(True)
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.hist(cryptos['ClosePrice'] , bins =30, color = 'green', edgecolor = 'black', alpha =0.8)
plt.title('Most Common Price Ranges')
plt.grid(True)
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

#### **Box plot show the distribution of prices**

```python
sns.boxplot(data= cryptos[['OpenPrice','HighPrice','LowPrice','ClosePrice']])
plt.show()
```

#### **Heatmap is ued to show correlation between multiple parameters**

```python
price_corr = cryptos[['OpenPrice', 'HighPrice', 'LowPrice', 'ClosePrice']].corr()
plt.figure(figsize=(10,5))
sns.heatmap(data = price_corr, annot = True, cmap = 'coolwarm', linewidths=0.5, linecolor='gray')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
plt.plot(cryptos['CloseTime'], cryptos['HighPrice'], label='High Price', color='green')
plt.plot(cryptos['CloseTime'], cryptos['LowPrice'], label='Low Price', color='red')
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('Bitcoin High and Low Prices Over Time')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(cryptos['OpenPrice'], cryptos['ClosePrice'], alpha=0.6, color='darkorange')
plt.xlabel('Open Price')
plt.ylabel('Close Price')
plt.title('Bitcoin: Open vs Close Price')
plt.grid(True, linestyle='--', alpha=0.3)
plt.tight_layout()
plt.show()
```
